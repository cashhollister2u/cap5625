# Part 2: Mean and Centering

### The mean vecor is:
$$ \mu=\frac{1}{n}\sum_{i=1}^{n} \vec{x}_{i}$$

### To center your data
$$ \bar{D}=D-\vec{1}.\mu^T$$

```python
D = D.to_numpy()
# Compute the mean vector
mean_vector=np.mean(D,axis=0)
# Center the data
D_centered=D-mean_vector
# Plot a scatter plot between the first two centered attributes
plt.scatter(D_centered[:,0],D_centered[:,1])
plt.grid()
```

# Part 3: Covariance Matrix


### Cov_matrix using inner product:
$$\Sigma=\frac{1}{n} \bar{D}^T \bar{D} $$

```python
# Implement covariance matrix w/ inner product 
n=D.shape[0]
cov_inner = np.dot(D_centered.T, D_centered)/n
cov_inner
#print(np.cov(D_centered, rowvar=False))
```

### Cov_matrix using outer product:
$$\Sigma=\frac{1}{n} \sum_{i=1}^{n} \bar{\vec{x}_{i}} \bar{\vec{x}_{i}}^T$$

```python
# implement covariance matrix w/ outer product
cov_outer = np.zeros((D_centered.shape[1], D_centered.shape[1]))
for i in range(D_centered.shape[0]):
    cov_outer += np.outer(D_centered[i], D_centered[i])

cov_outer /= D_centered.shape[0]
cov_outer
```

# Part 8: Eigen Decomposition 


### Formulate the eigen decomposition of the covariance matrix.
$$ \Sigma \vec{v}=\lambda \vec{v}$$

```python
def eigen_decomp(matrix):
    eigen_values, eigen_vectors = np.linalg.eigh(matrix)
    idx = np.argsort(eigen_values)[::-1] # get indices sorted in decending order
    eigen_values=eigen_values[idx] # redefine the variable with sorted indices
    eigen_vectors=eigen_vectors[:, idx] # redefine the variable with sorted indices

    v1=eigen_vectors[:,0] # first eigen vector
    print("Eigenvalues:", eigen_values)
    print("Eigenvectors:", eigen_vectors)
    print("First Eigenvector:", v1)

    # Project the data onto the first eigenvector
    projected_data = np.dot(D_centered, v1)[:,np.newaxis]*v1

    return projected_data
    

def plot_projection(projected_data):   
    # Plot the 1D projection.
    plt.scatter(D[:,0],D[:,1])
    plt.scatter(D_centered[:,0],D_centered[:,1],color='red')
    plt.scatter(projected_data[:,0],projected_data[:,1],color='green')
    plt.grid()
```

Do the same for the kernel matrices using their eigen decomposition.

```python
def rbf_kernel(x, y, gamma=1.0):
    """ Radial Basis Function (RBF) Kernel """
    # 1/2sigma^2 is labeled as gamma here
        # this is the gaussian kernel formula from my lecture notes
    return np.exp(-gamma * np.linalg.norm(x - y) ** 2)


def poly_kernel(x, y, degree=3):
    """ Polynomial Kernel """
    return np.dot(x, y) ** degree


def linear_kernel(x, y):
    """ Linear Kernel """
    return np.dot(x, y)

def euclidean_distance(x, y):
    """ Euclidean Distance """
    return np.linalg.norm(x - y)

def compute_kernel_matrix(data, kernel_type, **kwargs):
    """
    Compute Kernel Matrix Function

    PARAMETERS:
    - data: input data as an ndarray of shape (n_samples, n_features)
    - kernel_param: kernel parameter (e.g., gamma for RBF, degree for polynomial)

    RETURNS:
    - Kernel matrix.
    - n_samples, the number of samples in the input data (in this case, 200).
    """

    n_samples = data.shape[0]
    kernel_matrix = np.zeros((n_samples, n_samples))
    print(kernel_type)
    for i in range(n_samples):
        for j in range(n_samples):
            if (kernel_type=='rbf'):
                gamma = kwargs.get('gamma')
                kernel_matrix[i, j] = rbf_kernel(data[i], data[j], gamma=gamma)
            elif (kernel_type=='poly'):
                degree = kwargs.get('degree')
                kernel_matrix[i, j] = poly_kernel(data[i], data[j], degree=degree)
            elif (kernel_type=='linear'):
                kernel_matrix[i, j] = linear_kernel(data[i], data[j])
            elif (kernel_type=='euclidean'):
                kernel_matrix[i, j] = euclidean_distance(data[i], data[j])
            elif (kernel_type=='poly_rbf'):
                w1 = kwargs.get('w1')
                w2 = kwargs.get('w2')
                degree = kwargs.get('degree')
                gamma = kwargs.get('gamma')
                kernel_matrix[i, j] = w1*poly_kernel(data[i], data[j], degree=degree)+w2*rbf_kernel(data[i], data[j], gamma=gamma)

    return kernel_matrix, n_samples

def my_kpca(data, alpha, kernel_type='rbf', **kwargs):
    """
    Kernel Principal Component Analysis (KPCA) Function

    PARAMETERS:
    - data: input data as an ndarray of shape (n_samples, n_features)
    - alpha: percentage of the variance to retain (0-1, should be a decimal [e.g., 0.95])
    - kernel_type: type of kernel ('rbf', 'poly', or 'linear')
    - kernel_param: kernel parameter (e.g., gamma for RBF, degree for polynomial)

    SIDE NOTE:
    The following from the template code is no longer a parameter and is rather one of the returned items.
    - n_components: number of principal components
    This is because, in my version of my_kpca, I manually find the optimal value for n_components that retains the most data within the function.

    RETURNS:
    - Transformed data in the PCA space.
    - Reduced Eigen Vectors.
    - Mean.
    - k, the number of principle components.
    """

    # 1) computing the kernel matrix
    kernel_matrix, n_samples = compute_kernel_matrix(data, kernel_type, **kwargs)

    # 2) centering the kernel matrix
    one_n = np.ones((n_samples, n_samples)) / n_samples
    # split up for debugging
    mean2_1 = one_n.dot(kernel_matrix)
    mean2_2 = kernel_matrix.dot(one_n)
    mean2_3 = one_n.dot(kernel_matrix).dot(one_n)
    # combining
    mean2 = -(mean2_1) - (mean2_2) + (mean2_3)
    # centering
    centered_kernel_matrix = kernel_matrix + mean2

    # 3) finding the eigen values and eigen vectors
    eigen_values, eigen_vectors = np.linalg.eig(centered_kernel_matrix)

    # 4) sorting the eigen values and eigen vectors (descending order)
    idx = np.argsort(eigen_values)[::-1]
    eigen_values = eigen_values[idx]
    eigen_vectors = eigen_vectors[:, idx]

    # 5) Get the first eigenvector
    v1 = eigen_vectors[:, 0]

    # 6) Project onto the first eigenvector (1D projection)
    projection_1d = np.dot(centered_kernel_matrix, v1)[:,np.newaxis]*v1

    return projection_1d
```

```python
def my_kpca(data, alpha, kernel_type='rbf', **kwargs):
    """
    Kernel Principal Component Analysis (KPCA) Function

    PARAMETERS:
    - data: input data as an ndarray of shape (n_samples, n_features)
    - alpha: percentage of the variance to retain (0-1, should be a decimal [e.g., 0.95])
    - kernel_type: type of kernel ('rbf', 'poly', or 'linear')
    - kernel_param: kernel parameter (e.g., gamma for RBF, degree for polynomial)

    SIDE NOTE:
    The following from the template code is no longer a parameter and is rather one of the returned items.
    - n_components: number of principal components
    This is because, in my version of my_kpca, I manually find the optimal value for n_components that retains the most data within the function.

    RETURNS:
    - Transformed data in the PCA space.
    - Reduced Eigen Vectors.
    - Mean.
    - k, the number of principle components.
    """


    # 1) computing the kernel matrix
    kernel_matrix, n_samples = compute_kernel_matrix(data, kernel_type, **kwargs)


    # 2) centering the kernel matrix
    one_n = np.ones((n_samples, n_samples)) / n_samples
    # split up for debugging
    mean2_1 = one_n.dot(kernel_matrix)
    mean2_2 = kernel_matrix.dot(one_n)
    mean2_3 = one_n.dot(kernel_matrix).dot(one_n)
    # combining
    mean2 = -(mean2_1) - (mean2_2) + (mean2_3)
    # centering
    centered_kernel_matrix = kernel_matrix + mean2


    # 3) finding the eigen values and eigen vectors
    eigen_values, eigen_vectors = np.linalg.eig(centered_kernel_matrix)


    # 4) sorting the eigen values and eigen vectors (descending order)
    idx = np.argsort(eigen_values)[::-1]
    eigen_values = eigen_values[idx]
    eigen_vectors = eigen_vectors[:, idx]


    # EXTRA STEP) finding an optimal n_components value that retains a significant amount of variance
    var_ratio = eigen_values / np.sum(eigen_values)  # array that stores the var ratio for each eigen_value
    
    return var_ratio
```

# Part 10: Matrix Decomposition using SVD

### Formulate the SVD decomposition
- Centered data matrix:
    $$ A=U\Sigma V^T$$
- Covariance matrix:
    $$S=\frac{1}{n} U\Sigma\Sigma ^T U^T $$
- Linear Matrix:
    $$A^TA=V\Sigma\Sigma ^T V^T$$
- Polynomial Matrix:
    $$A^TA=(c+V\Sigma\Sigma ^T V^T)^{degree}$$ 
- RBF Matrix:
    $$A^TA = \exp\left\{ -\frac{\|(U\Sigma V^T)_i - (U\Sigma V^T)_j\|^2}{2\sigma^2} \right\}$$
- Euclidean Matrix:
    $$\|A^TA\| = \sqrt{A^TA + A^TA - 2(A^TA)}$$
- Polynomial/RBF Matrix:
    
$$A^TA = w_1 \cdot (c+V\Sigma\Sigma ^T V^T)^{degree} + w_2 \cdot \exp\left\{ -\frac{\|(U\Sigma V^T)_i - (U\Sigma V^T)_j\|^2}{2\sigma^2} \right\}$$


### Implement the SVD using Pythonâ€™s numpy.linalg.svd:

```python
def svd_decomp(D_centered):
    U, S, Vt = np.linalg.svd(D_centered)
    print(U, S, Vt)
    v1=Vt[:,0] # first eigen vector

    # Project the data onto the first eigenvector
    projected_data = np.dot(D_centered, v1)[:,np.newaxis]*v1
    
    return projected_data, U, S, Vt
```

```python
projected_data, U, S, Vt = svd_decomp(D_centered)
plot_projection(projected_data)
```

### Discuss how SVD relates to eigen decomposition

- SVD of a general matrix shares some of the similarities of the eigendecomposition of a square matrix. However, while eigendecomposition must be performed on a square matrix SVD can be applied to all matrices. Eigendecomposition operates from within the same vector space, where the same basis change is applied and then un-done. SVD expresses a change of basis in both domain and co-domain. Both bases are linked by the singular value matrix $\Sigma$. 

- Vectors in the eigendecomposition matrix P are not necessarily orthoginal. Vectors in SVD decomposition are orthoginal and represent rotations.

- Both Eigendecomposition and SVD are compositions of three linear mappings:
    - Change of basis in the domain.
    - Interdependent scaling of each new basis vector and mapping from domain to codomain. 
    - Change of basis in the codomain.

- Key difference: 
    - SVD has a domain and codomain that can be vector spaces of different dims as mentioned above (not required to be square)
